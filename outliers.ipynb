```python
upper_limit = data['AF Total'].mean() + 3*data['AF Total'].std()
lower_limit = data['AF Total'].mean() - 3*data['AF Total'].std()
```

```python
data['AF Total'] = np.where(
    data['AF Total']>upper_limit,
    upper_limit,
    np.where(
        data['AF Total']<lower_limit,
        lower_limit,
        data['AF Total']
    )
)
```
